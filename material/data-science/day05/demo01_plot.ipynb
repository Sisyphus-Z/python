## 绘制正弦、余弦函数图像

```python
# 从[-π, π] 拆出200个点
x = np.linspace(-np.pi, np.pi, 200)
sinx = np.sin(x)  # 矢量化方法    可以针对一组数据操作
# 绘制[-π, π]区间的： y=cos(x)/2
cosx = np.cos(x)/2

plt.plot(x, sinx, linewidth=2, linestyle='--', 
        color='dodgerblue', alpha=0.9 ,
        label=r'$y=sin(x)$')
plt.plot(x, cosx, linewidth=2, linestyle='-.',
        color='orangered', alpha=0.9,
        label=r'$y=\frac{1}{2}cos(x)$')

# 设置坐标轴的范围
# plt.xlim(0, np.pi)
# plt.ylim(0, 1)

# 设置x轴刻度文本
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
           [r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', 
            r'$\frac{\pi}{2}$', r'$\pi$'])

# 设置坐标轴  修改成平面直角坐标系
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
plt.yticks([-1, -0.5, 0.5, 1])

# 绘制特殊点  [π/2, 1] [π/2, 0] 
plt.scatter([np.pi/2, np.pi/2], [0, 1], marker='s', 
        s=150, edgecolor='green', facecolor='red', 
        zorder=3)

# 备注
plt.annotate(
    r'$[\frac{\pi}{2}, 1]$',
    xycoords='data',
    xy=(np.pi/2, 1),
    textcoords='offset points',
    xytext=(30, 20),
    fontsize=14,
    arrowprops=dict(
        arrowstyle='->', 
        connectionstyle='angle3')
)

plt.annotate(
    r'$[\frac{\pi}{2}, 0]$',
    xycoords='data',
    xy=(np.pi/2, 0),
    textcoords='offset points',
    xytext=(-50, -50),
    fontsize=14,
    arrowprops=dict(
        arrowstyle='->', 
        connectionstyle='angle3')
)

plt.legend()
plt.show()
```

## 网格式布局

```python
import matplotlib.gridspec as mg
gs = mg.GridSpec(3, 3)
# 合并第一行的前两列
plt.subplot(gs[0, :2])  
plt.text(0.5, 0.5, 1, size=30, va='center', ha='center')
plt.xticks([])
plt.yticks([])
plt.subplot(gs[:2, -1])  
plt.text(0.5, 0.5, 2, size=30, va='center', ha='center')
plt.xticks([])
plt.yticks([])
plt.subplot(gs[1, 1])  
plt.text(0.5, 0.5, 3, size=30, va='center', ha='center')
plt.xticks([])
plt.yticks([])
plt.subplot(gs[1:, 0])  
plt.text(0.5, 0.5, 4, size=30, va='center', ha='center')
plt.xticks([])
plt.yticks([])
plt.subplot(gs[-1, 1:])  
plt.text(0.5, 0.5, 5, size=30, va='center', ha='center')
plt.xticks([])
plt.yticks([])
plt.show()
```

## 散点图

```python
n = 200
x = np.random.normal(175, 5, n)
y = np.random.normal(65, 10, n)

# plt.scatter(x, y, s=60, color='dodgerblue', 
#      edgecolor='orangered', label='Samples')

plt.scatter(x, y, s=60, c=x, cmap='jet', label='Samples')
plt.legend()
```

```python

n = 1000
x = np.linspace(0, 8 * np.pi, n)  # 返回指定间隔上的等距数字

sin_y = np.sin(x)  # 计算sin函数值
cos_y = np.cos(x / 2) / 2  # 计算cos函数值

plt.figure('Fill', facecolor='lightgray')
plt.title('Fill', fontsize=20)
plt.xlabel('x', fontsize=14)  # x轴标签
plt.ylabel('y', fontsize=14)  # y轴
plt.tick_params(labelsize=10)  # 刻度
plt.grid(linestyle=':')

plt.plot(x, sin_y, c='dodgerblue', label=r'$y=sin(x)$')
plt.plot(x, cos_y, c='orangered', label=r'$y=\frac{1}{2}cos(\frac{x}{2})$')

# 填充cos_y < sin_y的部分
plt.fill_between(x, cos_y, sin_y, cos_y < sin_y, color='dodgerblue', alpha=0.5)
# 填充cos_y > sin_y的部分
plt.fill_between(x, cos_y, sin_y, cos_y > sin_y, color='orangered', alpha=0.5)

plt.legend()
plt.show()
```

## 柱状图

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
# 绘制12个月苹果的销量
x = np.arange(12)
apples = np.random.normal(80, 20, 12)
plt.bar(x-0.2, apples, 0.4, color='dodgerblue', label='apples')
oranges = np.random.normal(80, 20, 12)
plt.bar(x+0.2, oranges, 0.4, color='orangered', label='oranges')

plt.xticks(x, ['一月', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.show()
```
